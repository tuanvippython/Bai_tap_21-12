```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:/Users/ADMIN/Downloads/Data/healthcare_dataset.csv")
df
```

# Bài toán:dự đoán tiền viện phí (Billing Amount)

## Chọn ra những cột dữ liệu cần thiết để dự đoán

```python
df = df[["Age","Gender","Blood Type","Medical Condition","Date of Admission","Insurance Provider","Admission Type","Discharge Date","Medication","Test Results","Billing Amount"]]
df
```

## Kiểm tra các giá trị khác nhau của mỗi cột

```python
for column in df.columns:
    if not pd.api.types.is_numeric_dtype(df[column]):
        unique_values = df[column].unique()
        print(f'{column}: {unique_values}')
```

## Tính số ngày nằm viện và thêm vào dữ liệu

```python
from datetime import datetime
# Chuyển đổi các cột về định dạng datetime
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"])
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"])

# Tính số ngày nằm viện
df["Number of Days Hospitalized"] = (df["Discharge Date"] - df["Date of Admission"]).dt.days

# In ra DataFrame mới với cột "So ngay nam vien"
df
```

## Xóa ngày nhập viện và xuất viện

```python
df = df.drop(columns =["Date of Admission","Discharge Date"])
df
```

## Mã hóa dữ liệu

```python
# Chọn các cột cần mã hóa
columns_to_encode = ['Gender',"Blood Type", 'Medical Condition',"Insurance Provider", 'Admission Type', 'Medication',"Test Results"]

# Mã hóa các cột bằng dummy variables và chuyển đổi thành giá trị 0 và 1
dummies = pd.get_dummies(df[columns_to_encode], prefix=columns_to_encode, drop_first=True).astype(int)

# Thêm các cột đã mã hóa vào DataFrame gốc
df = pd.concat([df, dummies], axis=1)

# Loại bỏ các cột gốc đã được mã hóa
df = df.drop(columns=columns_to_encode)

# Hiển thị DataFrame sau khi thực hiện
df
```

## Chuẩn hóa dữ liệu

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Chọn các cột cần chuẩn hóa
columns = ["Age","Number of Days Hospitalized"]

# Chuẩn hóa dữ liệu và gán lại cho DataFrame
df[columns] = scaler.fit_transform(df[columns])

# In DataFrame sau khi chuẩn hóa
df
```

## Chia mô hình

```python
feature = df.iloc[:, [0] + list(range(2, len(df.columns)))]
target = df.iloc[:,1]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42)
```

## Tạo mô hình - huấn luyện - dự đoán

```python
from sklearn.linear_model import LinearRegression
# Tạo mô hình hồi quy
model = LinearRegression()

# Huấn luyện mô hình trên tập huấn luyện
model.fit(X_train, y_train)

# Dự đoán giá trên tập kiểm tra
y_predict = model.predict(X_test)
y_predict
```

## Kiểm tra độ chính xác

```python
from sklearn.metrics import mean_squared_error, r2_score

# Tính MSE
mse = mean_squared_error(y_test, y_predict)
print(f"Mean Squared Error: {mse}")

# Tính R2 score
r2 = r2_score(y_test, y_predict)
print(f"R-squared: {r2}")
```
